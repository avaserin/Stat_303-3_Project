```python
#| echo: false
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# Uploading data
data = pd.read_csv('boston.csv')

# Changing variable names
data.columns = ['crime_rate', 'zoned_land', 'nr_business', 'river', 'nitric_oxide', 'avg_rooms', 'prop_age', 
                'dist_emp', 'highway_acc', 'tax_rate', 'pt_ratio', 'race_prop', 'low_inc', 'median_val']

# Continuous Variable distributions
variable_dist = pd.DataFrame(columns=['Variable', 'Type', 'Missing', 'Unique', "Mean", "Std", "Min", "Max"])

for column in data.columns:
    dtype = str(data[column].dtype)
    missing = data[column].isna().sum()
    unique = data[column].nunique(dropna=False)
    mean = data[column].mean()
    std = data[column].std()
    min_val = data[column].min()
    max_val = data[column].max()
    variable_dist = variable_dist.append({'Variable': column, 'Type': dtype, 'Missing': missing, 'Unique': unique, 
                                          'Mean': mean, 'Std': std, 'Min': min_val, 'Max': max_val}, 
                                         ignore_index=True)

display(variable_dist)
```

Our team sought to optimize our predictive model's performance by using the root mean squared error (RMSE) as the metric of choice. RMSE is an effective tool for predicting Boston housing prices due to its ability to penalize large errors and its sensitivity to outliers. Both of these traits are critical, as substantial errors can lead to financial losses. Additionally, RMSE uses the same units as the response variable (in this case, thousands of dollars), which facilitates the interpretability of results for stakeholders. It should be noted that despite the existence of solutions on Kaggle for this dataset, we independently developed our models and sourced our dataset from an alternative location.

We started with a naive linear model as a baseline with an RMSE 5.214. Our modeling approach then included the creation and tuning of several models: MARS, bagged MARS, random forest, decision tree, bagged decision tree, ridge and lasso regressions, backward stepwise regression, XGBoost, and an ensemble model containing bagged decision trees, XGBoost, and random forest models, all with the goal of reducing the RMSE.
    
As we progressed, we encountered several challenges. We anticipated some, such as the long runtime associated with models like gradient boosting, which ultimately led to our decision to use the XGBoost model. However, other challenges were unexpected. We observed that some of the finely tuned models, such as the XGBoost and RandomForest and all tuning iterations for the unbagged decision tree models, exhibited poorer performance than we initially predicted.
